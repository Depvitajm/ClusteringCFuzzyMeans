Analisis Komponen Utama

```python
#Kovarian
def covariance_calculation(mean_subtracted):
    #mnghitung kovarian dari data standarisasi
    for col in mean_subtracted.columns:
        mean_subtracted[col] = mean_subtracted[col]-mean_subtracted[col].mean()
        
    return np.dot(mean_subtracted.T,mean_subtracted)/(len(mean_subtracted) - 1)

kovarian = covariance_calculation(datastandarisasi)
```

```python
print(kovarian)
```

```python
from numpy.linalg import eig
w,v=eig(kovarian)
```

```python
print (w)
```

```python
import matplotlib.pyplot as plt
plt.bar(["F" + str(i+1) for i in range(len(w))], w)
plt.bar_label(splot.containers[0], label_type='edge')
plt.title("Nilai Eigen")
plt.xlabel("Magnitude of the eigenvalue")
```
